```python
#L'obiettivo del notebook è creare un file xlsx che contenga le informazione di base di tutti i pazienti
import pandas as pd

patients=pd.read_json('./patientPuliti.json')
underlying_diseases = set()
basic_info=[]

#Creo la Set delle underlying sdiseases

for i,index in enumerate(patients):
    x = patients[index]['basic_info']['Underlying diseases']
    patient_i_underlying_diseases= x.split(', ')
    for _,disease in enumerate(patient_i_underlying_diseases):
        underlying_diseases.add(disease)

#Creo la struttura del dataframe

for _,basic in enumerate(patients['Patient 1']['basic_info']):
    if basic !='Underlying diseases':
        basic_info.append(basic)
basic_info.sort()


for _,disease in enumerate(underlying_diseases):
    basic_info.append(disease)

for _,r in enumerate(basic_info):
    print(r)

#Recupero i dati dal file json per l'aggiunta
data=[]
for i,index in enumerate(patients):
    patient_i = []
    for j,basic in enumerate(basic_info):
        if j<8:
            patient_i.append(patients[index]['basic_info'][basic])
        else:
            if patients[index]['basic_info']['Underlying diseases'].find(basic) >= 0:
                patient_i.append(1)
            else:
                patient_i.append(0) 

    data.append(patient_i)

writer = pd.ExcelWriter('patients_basic_info.xlsx')
dataframe = pd.DataFrame(data,columns=basic_info)
dataframe.to_excel(writer)
writer.save()
```
