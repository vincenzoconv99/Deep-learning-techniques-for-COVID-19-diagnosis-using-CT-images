```python
def get_new_diseases(old):
    new_classification_diseases = pd.read_excel('./disease_newclassification_EMA_Vin_Marco.xlsx')
    old_diseases = old.split(', ')
    new_diseases = []
    
    for index,row in enumerate(new_classification_diseases['Reported disease']):
        for _,old in enumerate(old_diseases):
            if old == row:
                new_diseases.append(new_classification_diseases['Comorbidity Classification'].get(index))
                
    return new_diseases

def get_organs(old):
    new_classification_diseases = pd.read_excel('./disease_newclassification_EMA_Vin_Marco.xlsx')
    old_diseases = old.split(', ')
    organs = []
    
    for index,row in enumerate(new_classification_diseases['Reported disease']):
        for _,old in enumerate(old_diseases):
            if old == row:
                organs.append(new_classification_diseases['Organ'].get(index))
                
    return organs
```
