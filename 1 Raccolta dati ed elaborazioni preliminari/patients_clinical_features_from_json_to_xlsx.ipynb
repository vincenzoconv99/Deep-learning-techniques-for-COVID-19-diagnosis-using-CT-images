```python
import pandas as pd

patients=pd.read_json('./patientPuliti.json')

clinical_features={}

for i,index_patient in enumerate(patients):
    keys = patients[index_patient]['cf'].keys()
    
    for j,index_key in enumerate(keys):
        for _,x in enumerate(patients[index_patient]['cf'][index_key]):
            normal_range = str(x['Normal range'])
            numeric_range = (normal_range.split(" ")[0].split('-'))
            
            if len(numeric_range) > 1:
                mins = (numeric_range[0])
                maxs = (numeric_range[1])
            else:
                if numeric_range[0][0]=='<':
                    mins= '-inf'
                    maxs=numeric_range[0].split('<')[1]
                if numeric_range[0][0]=='>':
                    mins= numeric_range[0].split('>')[1]
                    maxs= '+inf'
            if len(normal_range.split(" "))>1:        
                unit = normal_range.split(" ")[1]
            else:
                unit=''
            
            clinical_features[x['Abbreviation']]= {
                x['Abbreviation']+'_extendedname': x['Clinical feature'],
                x['Abbreviation']+'_min_normal_range': float(mins),
                x['Abbreviation']+'_max_normal_range': float(maxs),
                x['Abbreviation']+'_unit_normal_range': unit
            }
            
#File clinicalfeatures.xlsx
#contiene un indice con tutte le clinical feature

cf_for_excel=[]
abbreviations=clinical_features.keys()

print(len(abbreviations))

for _,key in enumerate(abbreviations):
    cf_for_excel.append([key, clinical_features[key][key+'_extendedname'], clinical_features[key][key+'_min_normal_range'], clinical_features[key][key+'_max_normal_range'], clinical_features[key][key+'_unit_normal_range']])

info = ['Abbreviation','Extended name', "Min Normal range", "Max Normal range", "Unit"]

writer = pd.ExcelWriter('clinical_features.xlsx')
dataframe = pd.DataFrame(cf_for_excel,columns=info)
dataframe.to_excel(writer)
writer.save()
```

```python
#Creazione del file patients_clinicalfeatures.xlsx
patients_cf=[]
info_for_patients=[]
for _,key in enumerate(abbreviations):
    info_for_patients.append(key+' Value')
    info_for_patients.append(key+' Value in range')
    info_for_patients.append(key+' Min Normal range')
    info_for_patients.append(key+' Max Normal range')
    info_for_patients.append(key+' Unit')


for i,_ in enumerate(patients):
    patients_cf.append({})
    patients_cf[i]['id'] = 'None'
    for _,key in enumerate(abbreviations):
        patients_cf[i][key+' Value'] = 'None'
        patients_cf[i][key+' Value in range'] = 'None'
        patients_cf[i][key+' Min Normal range'] = 'None'
        patients_cf[i][key+' Max Normal range'] = 'None'
        patients_cf[i][key+' Unit'] = 'None'
        
for i,index_patient in enumerate(patients):
    patients_cf[i]['id']= index_patient.replace(' ','_')
    keys = patients[index_patient]['cf'].keys()
    for j,index_key in enumerate(keys):
        for _,x in enumerate(patients[index_patient]['cf'][index_key]):
            normal_range = str(x['Normal range'])
            numeric_range = (normal_range.split(" ")[0].split('-'))
            
            if len(numeric_range) > 1:
                mins = (numeric_range[0])
                maxs = (numeric_range[1])
            else:
                if numeric_range[0][0]=='<':
                    mins= '-inf'
                    maxs=numeric_range[0].split('<')[1]
                if numeric_range[0][0]=='>':
                    mins= numeric_range[0].split('>')[1]
                    maxs= '+inf'
            if len(normal_range.split(" "))>1:        
                unit = normal_range.split(" ")[1]
            else:
                unit=''
            
            if x['Value'][0]!='<' and x['Value'][0]!='>':
                patients_cf[i][x['Abbreviation']+' Value']= (float(x['Value']) - float(mins))/(float(maxs)-float(mins))
                patients_cf[i][x['Abbreviation']+' Value in range']= x['Value in range']
                patients_cf[i][x['Abbreviation']+' Min Normal range']= mins
                patients_cf[i][x['Abbreviation']+' Max Normal range']= maxs
                patients_cf[i][x['Abbreviation']+' Unit']= unit
            else:
                patients_cf[i][x['Abbreviation']+' Value']= x['Value']
                patients_cf[i][x['Abbreviation']+' Value in range']= x['Value in range']
                patients_cf[i][x['Abbreviation']+' Min Normal range']= mins
                patients_cf[i][x['Abbreviation']+' Max Normal range']= maxs
                patients_cf[i][x['Abbreviation']+' Unit']= unit

df = pd.DataFrame.from_dict(patients_cf)
writer = pd.ExcelWriter('patient_clinical_features.xlsx')
df.to_excel(writer)
writer.save()
```
